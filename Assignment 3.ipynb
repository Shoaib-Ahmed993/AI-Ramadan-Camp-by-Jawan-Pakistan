```python
a = float(input("Enter first number:"))
b = float(input("Enter second number:"))
opr = input("Enter any operator:")
if (opr == '+'):
    ans = a+b
    print(f"{a} + {b} = {ans}")
elif (opr == '-'):
    ans = a-b
    print(f"{a} - {b} = {ans}")
elif (opr == '*'):
    ans = a*b
    print(f"{a} * {b} = {ans}")
elif (opr == '/'):
    if b != 0:
        ans = a/b
        print(f"{a} / {b} = {ans}")
    else:
        print("Error! Division by 0")
elif (opr == '**'):
    ans = a**b
    print(f"{a} ** {b} = {ans}")
else:
    print("You have entered invalid operator")
```

```python
numDictionary = {
    "a": 20,
    "b": 30,
    "c": 44.3
}
sum = 0
for check in numDictionary.values():
    if type(check) == int or type(check) == float:
        sum += check
print("Dictionary:",numDictionary)
print("Sum of numeric values:",sum)
```

```python
myList = ["Shoaib",20,12,12,31,"Shoaib",True,31,55,0.9,68,0.9,True]
duplicates = []
for item in myList:
    if myList.count(item) > 1 and item not in duplicates:
        duplicates.append(item)
print(f"Original List: {myList}")
print(f"Duplicate values are: {duplicates}")
```
