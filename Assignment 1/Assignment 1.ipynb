```python
English = int(input("Enter marks of English: "))
Urdu = int(input("Enter marks of Urdu: "))
Maths = int(input("Enter marks of Maths: "))
Islamiat = int(input("Enter marks of Islamiat: "))
Science = int(input("Enter marks of Science: "))

Obtained_marks = English + Urdu + Maths + Islamiat + Science
print("Your obtained marks: ", Obtained_marks)

Total_marks = 500
Percentage = int((Obtained_marks / Total_marks)*100)
print("Your percentage is: ",Percentage,"%")

if Percentage >=80:
    print("You've got A1 Grade")
elif Percentage >=70:
    print("You've got A Grade")
elif Percentage >=60:
    print("You've got B Grade")
elif Percentage >=50:
    print("You've got C Grade")
elif Percentage >=40:
    print("You've got D Grade")
elif Percentage <40:
    print("You're Failed")
else:
    print("Invalid")
```
