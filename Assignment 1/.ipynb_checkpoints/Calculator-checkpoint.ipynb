# Calculator App

```python

first_number = int(input("Enter first number: "))
second_number = int(input("Enter second number: "))
operator = input('Enter your operator: ')
if (operator == "+"):
    print("Answer is: ", first_number + second_number)
elif (operator == "-"):
    print("Answer is: ", first_number - second_number)
elif (operator == "*"):
    print("Answer is: ", first_number * second_number)
elif (operator == "/"):
    print("Answer is: ", first_number / second_number)
elif (operator == "%"):
    print("Answer is: ", first_number % second_number)
else:
    print("Invalid")
```
